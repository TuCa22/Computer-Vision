# Bài 1: Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1.1. Vẽ biểu đồ Histogram

Tạo ma trận

```python
img_mat = np.array([[0,3,1,2,4,1,2],
                   [3,3,1,2,5,6,1],
                   [2,1,1,1,3,1,3],
                   [5,3,2,2,1,3,4],
                   [0,0,0,3,7,0,1],
                   [1,1,3,3,2,1,2],
                   [3,2,1,1,2,3,0]])
print(img_mat)
```

Tính histogram ma trận

```python
histogram = np.zeros(8, dtype=int)  # Đếm số lượng pixel có giá trị từ 0 đến 7
for i in range(7):
    for j in range(7):
        histogram[img_mat[i, j]] += 1
```

Vẽ biểu đồ Histogram

```python
plt.bar(range(8), histogram)
plt.xlabel('Giá trị pixel')
plt.ylabel('Tần suất')
plt.title('Histogram ban đầu')
plt.show()
```

# 1.2. Cân bằng Histogram

Cân bằng Histogram

```python
total_pixels = np.sum(histogram)    # Tính tổng số lượng pixel
cumulative_sum = np.cumsum(histogram) / total_pixels    # Tính tỷ lệ tích lũy
new_mapping = (cumulative_sum * 7).astype(int)  # Tạo bản đồ ánh xạ mới

equal_mat = np.zeros((7, 7), dtype=int)  # Áp dụng cân bằng histogram cho ma trận
for i in range(7):
    for j in range(7):
        equal_mat[i, j] = new_mapping[img_mat[i, j]]   
```

Tính histogram sau khi cân bằng

```python
histogram_equalized = np.zeros(8, dtype=int)
for i in range(7):
    for j in range(7):
        histogram_equalized[equal_mat[i, j]] += 1 
```

Ma trận ban đầu và ma trận đã cân bằng histogram

```python
print("Ma trận ban đầu:")
print(img_mat)
print("\nMa trận đã cân bằng histogram:")
print(equal_mat)
```

Biểu đồ Histogram ban đầu và sau khi cân bằng

```python
plt.subplot(1, 2, 1)    # Biểu đồ histogram ban đầu
plt.bar(range(8), histogram)
plt.title('Histogram Ban đầu')
plt.xlabel('Pixel Value')
plt.ylabel('Số lượng pixel')

plt.subplot(1, 2, 2)    # Biểu đồ histogram cân bằng
plt.bar(range(8), histogram_equalized)
plt.title('Histogram Sau khi cân bằng')
plt.xlabel('Pixel Value')
plt.ylabel('Số lượng pixel')

plt.tight_layout()
plt.show()
```
