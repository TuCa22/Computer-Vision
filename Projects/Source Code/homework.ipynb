```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread('D:\Programming\Computer Vision\Projects\Images\saturn.jpg')
plt.imshow(cv2.cvtColor(image1,cv2.COLOR_BGR2RGB))
```

```python
image2 = cv2.imread("D:\Programming\Computer Vision\Projects\Images\phi_hanh_gia.jpg")
plt.imshow(cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)) 
```

```python
image1_gray = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
```

```python
# Đảm bảo cùng kích thước cho hai ảnh
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Tỷ lệ trọng số của ảnh thứ nhất và ảnh thứ hai trong image blending
alpha = 0.5
beta = 1 - alpha

# Thực hiện image blending
blended_image = cv2.addWeighted(image1, alpha, image2, beta, 0)
```

```python
plt.imshow('Weighted Image', blended_image)
```
